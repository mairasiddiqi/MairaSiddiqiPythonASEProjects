## Find the Most Recent 10 Albums

```python
data = pd.read_csv('albumlist.csv')

# Convert 'Year' column to numeric (to avoid issues if it's stored as a string)
data["Year"] = pd.to_numeric(data["Year"], errors="coerce")

# Sort by 'Year' in descending order (most recent first)
sorted_data = data.sort_values(by="Year", ascending=False)

# Get the most recent 10 albums as a list of dictionaries
recent_albums = sorted_data.head(10).to_dict(orient="records")

# Print the list
print(recent_albums)
```

## Visualize in Line Graph Any Comparison of Top 5 Anything

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("albumlist.csv")

df["Year"] = pd.to_numeric(df["Year"], errors="coerce")

df = df.dropna(subset=["Year"])

df["Year"] = df["Year"].astype(int)

top_artists = df["Artist"].value_counts().nlargest(5).index

top_5_df = df[df["Artist"].isin(top_artists)]

plt.figure(figsize=(10, 5))

for artist in top_artists:
    artist_data = top_5_df[top_5_df["Artist"] == artist]
    album_counts = artist_data.groupby("Year").size()
    plt.plot(album_counts.index, album_counts.values, marker="o", label=artist)

plt.xlabel("Year")
plt.ylabel("Number of Albums Released")
plt.title("Top 5 Artists: Album Releases Over Time")
plt.legend()
plt.grid(True)

plt.show()
```
