## Barchart usingplt.bar()


```python
import pandas as pd
import matplotlib.pyplot as plt

categories = ["Category A", "Category B", "Category C", "Category D"]
values =[23, 45, 56, 78]

#Creating the bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Examples')
plt.show()

```

## Barchart Using Dataframe's plot() Method

```python
import pandas as pd

df = pd.DataFrame({
    'city': ['new york', 'los angeles', 'chicago', 'houstan', 'phoenix'],
    'buisness': [500, 400, 300, 500, 1000]
})

df.set_index('city')['buisness'].plot(kind='bar', figsize=(10, 6), color='green', edgecolor='black', title='Number of Buisnesses Per City')

plt.ylabel('Number of Buisnesses')
plt.xticks(rotation=45)
plt.show()
```

## Plotting a Pivot Table

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'category_0': ['Bars', 'Resturants', 'Cafes', 'Fastfood'],
    'stars': [4.2, 3.8, 4.5, 2.8]
})

# Creating a pivot table to get the average star rating by category
pivot_state_cat = pd.pivot_table(df, values='stars', index=['category_0'], aggfunc='mean')

# Plotting the average stars per category
pivot_state_cat.plot(kind='line', figsize=(10, 6), color='purple', title='Average Stars Rating For Food Places')

plt.xlabel('Categories')
plt.ylabel('Average stars per category')
plt.xticks(rotation=45)
plt.show()
```
