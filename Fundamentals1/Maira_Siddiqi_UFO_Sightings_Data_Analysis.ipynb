# UFO Sightings Data Analysis

This Jupyter notebook outlines steps to load and analyze a dataset containing UFO sightings.

## Problem 1: Load and Read Data

- **Task**: Import the `csv` module, load and read the UFO sightings dataset from `ufo-sightings.csv` using a `DictReader` in a `with` statement.
- **Output**:
  - Print the field names of the dataset.
  - Iterate over the reader to store the data into a list named `ufosightings`.


```python
# Write your code here

# UFO Sightings Data Analysis

import csv

# Open and read the dataset using DictReader
with open('ufo-sightings.csv', mode='r') as file:
    # Create a DictReader object to read the data
    reader = csv.DictReader(file)

    # Print the field names
    print("Field Names:", reader.fieldnames)

    # Store data into a list
    ufosightings = []
    for row in reader:
        ufosightings.append(row)

    # Optional: Print first few entries of the ufosightings list to check the data
    print("First few UFO Sightings:")
    print(ufosightings[:5])



```

## Problem 2: Total Sightings

- **Task**: Calculate the total number of UFO sightings.
- **Output**: Print the total number of sightings stored in a variable `ufosightings_count`.

```python
# Write your code here


# Calculate the total number of UFO sightings
ufosightings_count = len(ufosightings)

# Print the total number of sightings
print(f"Total number of UFO sightings: {ufosightings_count}")




```

## Problem 3: Sightings in the US

- **Task**: Determine how many UFO sightings occurred in the US.
- **Output**: Store US sightings in a list named `sightings_us` and print the count.


```python
# Write your code here

# Filter the sightings that occurred in the US
sightings_us = [sighting for sighting in ufosightings if sighting['country'] == 'us']

# Calculate the number of sightings in the US
us_sightings_count = len(sightings_us)

# Print the number of sightings in the US
print(f"Total number of UFO sightings in the US: {us_sightings_count}")
```

## Problem 4: Fireball Sightings Over Ten Seconds

- **Task**: Find "fireball" sightings that lasted more than ten seconds in the US.
- **Details**:
  - Cast the duration to a float.
  - Filter sightings where `shape` is "fireball" and duration is greater than 10 seconds.
- **Output**: Print the datetime and state of each sighting.


```python
# Write your code here

# Filter fireball sightings in the US that lasted more than 10 seconds
fireball_sightings_over_10s = [
    sighting for sighting in sightings_us
    if sighting['shape'].lower() == 'fireball' and float(sighting['duration (seconds)']) > 10
]

# Print the datetime and state of each filtered sighting
for sighting in fireball_sightings_over_10s:
    print(f"Datetime: {sighting['datetime']}, State: {sighting['state']}")

```

## Problem 5: Longest Fireball Sighting

- **Task**: Sort the fireball sightings by duration and determine the longest sighting.
- **Output**:
  - Sort the list `fball` by duration in descending order.
  - Print the datetime and duration of the longest sighting.


```python
# Write your code here

# Filter fireball sightings in the US that lasted more than 10 seconds
fball = [
    sighting for sighting in sightings_us
    if sighting['shape'].lower() == 'fireball' and float(sighting['duration (seconds)']) > 10
]

# Sort the fireball sightings by duration in descending order
fball_sorted = sorted(fball, key=lambda x: float(x['duration (seconds)']), reverse=True)

# Print the datetime and duration of the longest fireball sighting
if fball_sorted:
    longest_sighting = fball_sorted[0]
    print(
        f"Longest Fireball Sighting: Datetime: {longest_sighting['datetime']}, Duration: {longest_sighting['duration (seconds)']} seconds")
else:
    print("No fireball sightings over 10 seconds.")
```

## Problem 6: State with Longest Fireball Sighting

- **Task**: Identify the state that had the longest lasting fireball sighting.
- **Output**: Print the state.


```python
# Write your code here

# Sort the fireball sightings by duration in descending order (same as in Problem 5)
fball_sorted = sorted(fball, key=lambda x: float(x['duration (seconds)']), reverse=True)

# Identify the state of the longest lasting fireball sighting
if fball_sorted:
    longest_sighting_state = fball_sorted[0]['state']
    print(f"State with the longest lasting fireball sighting: {longest_sighting_state}")
else:
    print("No fireball sightings over 10 seconds.")



```

## Problem 7: Filter Insignificant Sightings

- **Task**: Assume any sighting of zero seconds is insignificant. Filter these records out.
- **Output**: Find and print the shortest sighting duration among the significant records.


```python

# Function to safely convert the duration to float, handling invalid entries
def safe_float(value):
    try:
        return float(value)
    except ValueError:
        return None  # Return None if conversion fails


# Filter out sightings with zero seconds duration and invalid durations
significant_sightings = [
    sighting for sighting in ufosightings
    if safe_float(sighting['duration (seconds)']) and safe_float(sighting['duration (seconds)']) > 0
]

# Find the shortest duration among the significant sightings
if significant_sightings:
    shortest_sighting = min(significant_sightings, key=lambda x: safe_float(x['duration (seconds)']))
    print(f"Shortest significant sighting duration: {safe_float(shortest_sighting['duration (seconds)'])} seconds")
else:
    print("No significant sightings (all durations are zero or invalid).")
```

## Problem 9: Top 10 States with the most UFO sightings

- **Task**: Create a line graph of the following
  - Top 10 of the US states with the most UFO sightings
  

```python
# Write your code here

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ufo-sightings.csv', low_memory=False)


state_counts = df['state'].value_counts()


top_10_states = state_counts.head(10)


plt.figure(figsize=(10, 5))
plt.plot(top_10_states.index, top_10_states.values, marker='o', linestyle='-', color='b')

plt.xlabel("States")
plt.ylabel("Number of Sightings")
plt.title("Top 10 States with the Most UFO Sightings")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()


```

## Problem 10: Comparing three years (2010, 2011, 2012) top 10 signthings

- **Task**: Create a line graph of the following
  - Top 10 of the US states with the most UFO sightings
  

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ufo-sightings.csv', low_memory=False)

df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df['year'] = df['datetime'].dt.year

df_filtered = df[df['year'].isin([2010, 2011, 2012])]
state_counts = df_filtered.groupby(['year', 'state']).size().reset_index(name='count')

# For each year, get the top 10 states without using apply
top_10_states_per_year = state_counts.sort_values(['year', 'count'], ascending=[True, False]) \
    .groupby('year').head(10).reset_index(drop=True)

plt.figure(figsize=(12, 6))
colors = {2010: 'b', 2011: 'g', 2012: 'r'}

for year, data in top_10_states_per_year.groupby('year'):
    plt.plot(data['state'], data['count'], marker='o', linestyle='-', color=colors[year], label=f"{year}")

plt.xlabel("States")
plt.ylabel("Number of Sightings")
plt.title("Top 10 US States with the Most UFO Sightings (2010-2012)")
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.grid(True)

plt.show()
```

## UFO Sightings Heatmap by Time of Day and Month

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('ufo-sightings.csv', low_memory=False)

# Convert 'datetime' to datetime type
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')

# Extract the hour and month from the 'datetime' column
df['hour'] = df['datetime'].dt.hour
df['month'] = df['datetime'].dt.month

# Create a pivot table: Rows = month, Columns = hour, Values = count of sightings
heatmap_data = df.groupby(['month', 'hour']).size().unstack(fill_value=0)

# Plotting the heatmap using imshow
plt.figure(figsize=(12, 6))
plt.imshow(heatmap_data, cmap="YlGnBu", aspect='auto', interpolation='nearest')

# Add color bar
cbar = plt.colorbar()
cbar.set_label('Number of Sightings')

# Set axis labels
plt.title('UFO Sightings Heatmap by Time of Day and Month')
plt.xlabel('Hour of Day')
plt.ylabel('Month of Year')

# Customize ticks
plt.xticks(np.arange(24), [str(i) for i in range(24)])  # Hour ticks (0-23)
plt.yticks(np.arange(12), [str(i + 1) for i in range(12)])  # Month ticks (1-12)

# Display the plot
plt.show()
```
